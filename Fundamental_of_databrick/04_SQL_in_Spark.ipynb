```python
#Creating own dataframe
from pyspark.sql import functions as f, types as t

data = [
    ("2017-01-01", 32.0, 6.0, "Rain"),
    ("2017-01-04", None, 9.0, "sunny"),
    ("2017-01-05", 22.0, None, "Snow"),
    ("2017-01-06", 19.0, 9.0, None),
    ("2017-01-07", 19.0, None, "Rain"),
    ("2017-01-08", None, None, "Sunny"),
    ("2017-01-09", 0.0, 5.0, "Rain"),
    ("2017-01-10", None, None, None),
    ("2017-01-11", 0.0, 7.0, "Rain"),
    ("2017-01-12", 7.0, 2.0, "Snow"),
    ("2017-01-13", 4.0, 1.0, "sunny"),
    ("2017-01-14", 4.0, 0.0, "Rain")
]

schema = t.StructType([
    t.StructField("date", t.DateType()),
    t.StructField("high", t.DoubleType()),
    t.StructField("low", t.DoubleType()),
    t.StructField("weather", t.StringType())
])

df = spark.createDataFrame(data, schema)
display(df)
```

```python
schema1 = "day string, temperature double, windspeed double, event string"

df = spark.createDataFrame(data, schema1)
df = df.withColumn("day", f.to_date("day", "yyyy-MM-dd")) #normalize to DateType
display(df)
```

```python
df.createOrReplaceTempView("weather") #It is for this session
# df.createGlobalTempView("global_weather") #It is for all session
```
