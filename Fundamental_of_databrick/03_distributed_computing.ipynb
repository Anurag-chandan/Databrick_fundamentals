### Distributed computing

* **Distributed computing :-** It is the process of splitting a large computational task into smaller subtasks and executing them in parallel across multiple machines that work together as a single system.

* Hadoop is framework that allows for the distributing processing of large datasets across cluster of computers using simple programming models.
* **Hadoop is a distributed computing framework.**

Disadvantage of Hadoop :-
* Hadoop wask disk heavy and slow.


### Apache Spark
* Apache Spark is an open-source, distributed computing engine designed to process large-scale data quickly and efficiently.
* A distributed compute engine that abstracts that complexity of parallelism, letting developers focus on business logic.
* It is computing engine like hadoop.

Difference b/w Hadoop and Apache Spark :-
1. Spark process data in-memory while hadoop is disk heavy.


* Apache spark is faster than Hadoop.
