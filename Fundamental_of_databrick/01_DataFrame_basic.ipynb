## Column Filtering

```python
df1.select("title","budget","revenue","imdb_rating","industry")
#Since, it is immutable in nature, it will not change the original dataframe
#To change the original dataframe
df1_trimmed = df1.select("title","budget","revenue","imdb_rating","industry")
# df1_trimmed.show()
display(df1_trimmed)
```

```python
#To get the distinct values of a column
df1.select("industry").distinct().show()
```

```python
#Using truncate parameter in show() method
#If truncate = True, it will ommit the longer values in column.
df1_trim = df1.select("title","imdb_rating","industry","release_year")
df1_trim.show(3,truncate=False)
```

## Row Filtering

```python
#To get movies which are release b/w 2000 and 2010
df_filter = df1_trim.filter((df1_trim.release_year >= 2000) & (df1_trim.release_year <= 2010))
# df_filter = df1_trim.filter((df1_trim.release_year >= 2000) | (df1_trim.release_year <= 2010))
#df_filter = df1_trim.filter(df1_trim.release_year.isin([2000,2010]))
#df_filter = df1_trim.filter((df1_trim["release_year"] >= 2000) & (df1_trim["release_year"] <= 2010))
# df_filter = df1_trim.filter(df1_trim["release_year"].isin([2000,2010]))

#we can also use between() method
# df_filter = df1_trim.filter(df1_trim.release_year.between(2000,2010))
df_filter.show(3,truncate=False)
# display(df_filter)
```

```python
df_filter = df1_trim.filter((df1_trim["release_year"] >= 2000) & (df1_trim["release_year"] <= 2010))
df_filter.show(3,truncate=False)
```

```python
#Using col function
from pyspark.sql.functions import col
df_filter = df1_trim.filter((col("release_year") >= 2000) & (col("release_year") <= 2010))
df_filter.show(3,truncate=False)
```

```python
#using between
from pyspark.sql.functions import col
df_filter = df1_trim.filter(col("release_year").between(2000,2010))
df_filter.show(3,truncate=False)
```

## Creating columns

```python
from pyspark.sql.functions import round, col
df1 = df1.withColumn("profit",col("revenue")-col("budget"))
display(df1)
df1_round = df1.withColumn("profit",round(col("profit"),2))
display(df1_round)
```

```python
#To change column name
df1 = df1.withColumnRenamed("revenue","total_revenue")
df1.printSchema()
```
