* To see QueryPlan in SQL we use explain() and in PySpark we use ".explain()".


```python
df = spark.table("codebasics.fundamental.movies")
display(df)
```

```python
from pyspark.sql import functions as f

df_narrow = df.select("title","studio","imdb_rating").filter(f.col("release_year")>2000)
display(df)
```

```python

#Using .explain(extended=True)
df_narrow.explain(extended=True)
#Using .explain(extended="codegen")
# df_narrow.explain(extended="codegen")
```
