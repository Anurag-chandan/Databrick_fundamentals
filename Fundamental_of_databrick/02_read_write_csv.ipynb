```python
#To read the data from volume
df = spark.read.format("csv").\
    option("header", "true").\
    option("inferSchema", "true").\
    load("/Volumes/codebasics/fundamental/codebasics_volume/orders.csv")
display(df)
```

```python
#To add Schema to data
from pyspark.sql import functions as F
from pyspark.sql.types import *
schema = StructType([
    StructField('order_date', DateType()),
    StructField('country', StringType()),
    StructField('order_id', IntegerType()),
    StructField('product', StringType()),
    StructField('qty', IntegerType()),
    StructField('price', DoubleType())
])
df1 = spark.read.schema(schema).csv("/Volumes/codebasics/fundamental/codebasics_volume/orders.csv")
display(df1)
```

```python
#To write csv into parquet file
df.write.mode("overwrite").parquet("/Volumes/codebasics/fundamental/codebasics_volume/orders.parquet")
#To read parquet file
df_parquet = spark.read.parquet("/Volumes/codebasics/fundamental/codebasics_volume/orders.parquet")
display(df_parquet)
```
